# Splitting the data into training and testing datasets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(data_cancer_modified,data_cancer_labels,test_size=0.25,random_state=1)
```

# Model Learning

I will be using three models which are Decision Tree, Logistic Regression and Naive Bayes for classification of cancer dataset.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score,mean_squared_error
from sklearn.metrics import confusion_matrix
```

```python
model1 = LogisticRegression()
```

```python
model1_fit = model1.fit(X_train,y_train)
```

```python
y_pred_model1 = model1.predict(X_test)
```

```python
model1_score = accuracy_score(y_pred_model1,y_test)
model1_mean_square = mean_squared_error(y_pred_model1,y_test)
model1_confusion_matrix = confusion_matrix(y_pred_model1,y_test)
```

```python
model1_score
```

```python
model1_mean_square
```

```python
model1_confusion_matrix
```

```python
model2 = GaussianNB()
```

```python
model2_fit = model2.fit(X_train,y_train)
```

```python
y_pred_model2 = model2.predict(X_test)
```

```python
model2_score = accuracy_score(y_pred_model2,y_test)
model2_mean_square = mean_squared_error(y_pred_model2,y_test)
model2_confusion_matrix = confusion_matrix(y_pred_model2,y_test)
```

```python
model2_score
```

```python
model2_mean_square
```

```python
model2_confusion_matrix
```

```python
model3 = DecisionTreeClassifier()
```

```python
model3_fit = model3.fit(X_train,y_train)
```

```python
y_pred_model3 = model3.predict(X_test)
```

```python
model3_score = accuracy_score(y_pred_model3,y_test)
model3_mean_square = mean_squared_error(y_pred_model3,y_test)
model3_confusion_matrix = confusion_matrix(y_pred_model3,y_test)
```

```python
model3_score
```

```python
model3_mean_square
```

```python
model3_confusion_matrix
```

```python
#visualising the accuracy plot
plt.scatter(1,model1_score)
plt.scatter(1.05,model2_score)
plt.scatter(1.1,model3_score)
plt.legend(['Logistic Accuracy','Bayes Accuracy','Tree Accuracy'])
plt.title('Accuracy')
plt.ylabel('Accuracy')
plt.show()

#visulising the error plot
plt.scatter(1,model1_mean_square)
plt.scatter(1.05,model2_mean_square)
plt.scatter(1.1,model3_mean_square)
plt.legend(['Logistic Mean Square','Bayes Mean Square','Tree Mean Square'])
plt.title('Mean Squared Error')
plt.show()
```
